## Markdown / Latex 설명

## Markdown <br>
웹 문서 작성을 위해서 html을 배워야한다. 이러한 수고를 덜기 위해 기본적인 html 지식이 없어도 글을 작성하는데 어려움이 
없게 해주는 것이 **Markdown**이다. 그럼 word로 작성할 수도 있고, excel로 작성할 수도 있는데 왜 굳이 Markdown을 배워야 할까? <br>

내 생각은 웹을 통해 배포하는 자료가 기존 다른 자료 형식과의 차이점이 있기 때문이다. <br>
기본적인 base 자료를 만들기 위해서 주로 우리는 excel과 word를 함께 사용한다. 이 자료들은 단점이 존재하는데 바로 누군가에게 전달
을 할 때에는 효용이 매우 떨어진다는 측면이다. (*이를 보완하기 위한 자료가 ppt이다.*) <br>

웹에서 어떤 설명하는 자료를 받아서 볼 때 excel이나 word로 본 기억이 거의 없을 것이다. 이렇게, 어떤 해석없이 자료를 raw 그대로 전달하면 
해석하기가 너무 어렵고 이를 나타낸 ppt를 보면 상세한 내용은 알기 어려워서 일 것이다. <br>
하지만 blog에서 설명한 자료를 봤을 때 상세한 내용에 그림이 덧붙여져있어 이해하기도 쉬우며 보기에도 편하다. 이게 내가 생각하는 Markdown을 배워서 웹에 자료를 배포해야 하는 이유이다. <br>
웹에 문서를 띄워서 보기 편하게 해주면서(*파일 클릭을 안 해도 된다.*) 다른 페이지에 있는 내용들을 자유롭게 끌어다쓸 수 있다. (*옛날과 달리 특정 책에서만 발췌했을 때 해당 문서는 이해하기도 어렵고 필요한 내용이 많이 부족하게 된다.*) <br>

이런 장점들로 인해 blog를 통해 많은 사람들이 최신 기술에 대한 동향이나 논문에 대한 리뷰 등 좋은 자료들을 많이 남기고 있고 웹의 특성상
다양한 피드백을 받아 자료를 더 낫게 만들고 있다.

### Markdown 문법
대표적인 문법 몇 가지만 소개하고 더 필요한 내용은 아래 참고 자료 내용을 보도록 하겠다. <br>
아래 내용은 별거없다. 제목 크기를 다르게 하고, 들여쓰고 목록의 경우 순서를 매기고, 코드 작성이 필요한 경우 해당 구간을 
다르게 표시해주고, 수평선을 그어 구역을 나누어주고, 강조가 필요한 단어를 표시해주고, 그림이나 url이 필요한 경우 삽입해준다.

우리가 보통 문서를 작성할 때 필요한 내용이며 더 필요한 부분이 있을 경우 html 문법이나 웹 검색을 통해 알아보도록 하자.


***

## Latex
수학적인 내용을 기호로 표현할 때 쓰는 언어이다. <br>
자연과학, 공학 분야를 공부하게 되면 자연스럽게 수식을 많이 사용하게 되는데 이를 편리하게 작성할 수 있도록 도와주는 언어이다.
수식 관련 대표적인 공용 언어라고 생각하면 되며 처음에 사용했을 때 어려움은 있겠지만 **Markdown**과 함께 사용했을 때 매우 큰 장점을 갖는다.
장점을 나열하자면 첫번째로는 거의 모든 웹에서의 문서가 이 언어를 지원한다는 점이며 (*한번 배워두면 영원히 쓸 수 있다.*) <br>
두번째로는 생각하는 모든 표현을 나타낼 수 있다는 점이다. <br>

첫번째 내용 같은 경우에는 markdown을 배우는 이유기도 하고 github을 사용하는 이유기도 하다. 많은 사람들이 사용하면 할 수록 대표성을 띄게 되어 관련 업무를 함께 할 때 필요한 필수 도구가 되어 버린다. <br>
두번째 내용이 중요한데 자유도가 높다는 점이다. 내가 표현하고자 하는 것들을 가장 자유롭게 표현할 수 있어야 하는데 **latex**는 완벽하게 지원한다. 

### Latex 문법

사실 Latex는 다양한 기호가 표시되는 언어를 아는게 문제이므로 어디에서 찾아서 쓸 수 있을지 웹 사이트를 알려주고 어떻게 사용하는 지 몇 가지
예시만 들도록 하겠다.

**추천 사이트** <br>
1) [기호 검색](http://detexify.kirelabs.org/classify.html) : 필요한 기호를 그리면 매칭시켜서 찾아준다.  <br>
2) [Documentation](http://www.onemathematicalcat.org/MathJaxDocumentation/MathJaxKorean/TeXSyntax_ko.html) : 모든 Tex에 대한 정보를 담고 있는 문서이다. 위 사이트에서 찾다가 어려운 점들은 해당 사이트에서 찾으면 편리하다.  

**기본 작성 tip** <br>
- 기본적으로 Latex는 **$$**로 시작하고 끝난다. <br>
- 변수를 묶고 싶을 경우엔 **{}**로 포함해준다. <br>
- 띄어쓰기는 기본적으로 무시하며 문자 뒤에 **\**를 붙이면 한칸 띄어쓰기가 된다.
- 분수같은 경우 표현 방식이 헷갈릴 수 있는데 그냥 순서대로 쓰면 된다. 분모/분자 구역만 잘 나누어주면 된다.
- **_, ^**를 제일 많이 쓴다. _는 아래 쪽 첨자, ^는 위 쪽 첨자를 나타내는 기호라고 생각하면 된다. 

```
예시)
$$y_i = w_i*x_i$$
$$P_{ij} = P(w_j\lvert w_i) = {X_{ij} \above 1pt X_i}$$
```
$$y_i = w_ix_i$$
$$P_{ij} = P(w_j\lvert w_i) = {X_{ij} \above 1pt X_i}$$
