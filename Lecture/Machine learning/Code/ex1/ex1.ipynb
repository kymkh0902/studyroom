### <font color="blue">2. Linear regression with one variable</font>
 - **Hypothesis** : $h_{\theta}(x)=\theta_{0}+\theta_{1}x$
 - Parameters : $\theta_0,\theta_1$
 - **Cost Function** : $J(\theta_{0},\theta_{1})=\frac{1}{2m}\sum^m_{i=1}(h_{\theta}(x^{(i)})-y^{(i)})^2$
 - **Goal** : minimize $J(\theta_0,\theta_1)$  and $\theta_0,\theta_1$
 - **Gradient descent algorithm** : $\theta_j:=\theta_j-\alpha\frac{\partial}{\partial\theta_j}J(\theta_0,\theta_1)$ (for $j=0$ and $j=1$)<br>
  *(**Simultaneous Update** = 일시적으로, temp0, temp1을 모두 **일시적**으로 계산해야 한다)*<br>
  *(또한, $\alpha$(learning rate)는 항상 **양수**의 값을 가진다)*<br><br>
 - $\frac{\partial}{\partial\theta_j}J(\theta_0,\theta_1)=\frac{\partial}{\partial\theta_j}\frac{1}{2m}\sum^m_{i=1}(h_\theta(x^{(i)}-y^{(i)})^2=\frac{\partial}{\partial\theta_j}\frac{1}{2m}\sum^m_{i=1}(\theta_0+\theta_1x^{(i)}-y^{(i)})^2$<br><br> 
 - $j=0:\frac{\partial}{\partial\theta_0}J(\theta_0,\theta_1)=\frac{1}{m}\sum^m_{i=1}(h_\theta(x^{(i)}-y^{(i)})$<br><br>
 - $j=1:\frac{\partial}{\partial\theta_1}J(\theta_0,\theta_1)=\frac{1}{m}\sum^m_{i=1}(h_\theta(x^{(i)}-y^{(i)})$ $\cdotp x^{(i)}$<br><br>
 - $\theta_0:=\theta_0-\alpha\frac{1}{m}\sum^m_{i=1}(h_\theta(x^{(i)}-y^{(i)})$<br>
 - $\theta_1:=\theta_1-\alpha\frac{1}{m}\sum^m_{(i=1)}(h_\theta(x^{(i)}-y^{(i)})$ $\cdotp x^{(i)}$<br>
  **(update $\theta_0$ and $\theta_1$ simultaneously)**<br><br>

### <font color="blue">3. Linear Algebra review
 - **Matrix multiplication properties**<br>
 1) **not commutative** (교환법칙 성립이 안됨) - $A\times B\ne B\times A$<br>
 2) **Associate** (결합법칙) - $A\times(B\times C)=(A\times B)\times C$<br>
 3) **Identity Matrix** - $A\cdot I=I\cdot A=A$
 - **Matrix inverse** : $AA^{-1}=A^{-1}A=I$<br>
 *(Matrices that don't have an inverse are "singular(단일형)" or "degenerate")*
 - **Matrix Transpose** : $B=A^{T}, B_{ij}=A_{ji}$<br>
 *( ex) if : $A=\begin{bmatrix}2&3\cr4&5\cr6&7\end{bmatrix}$, then $A^{T}=\begin{bmatrix}2&4&6\cr3&5&7\end{bmatrix}$)*

## Programming Exercise 1: Linear Regression

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rel_error(x, y):
    """ returns relative error """
    return np.max(np.abs(x - y) / (np.maximum(1e-8, np.abs(x) + np.abs(y))))
```

### 2 Linear regression with one variable

```python
datafile = 'data/ex1data1.txt'
##############################################################################
### np.loadtxt를 사용해서 datafile을 불러오기.                                 ###
### 조건 :구분자는 "," 이고 데이터 범위는 column 기준 0, 1번째, transpose : True   ###
##############################################################################

### 코드 시작 ### 
cols = np.loadtxt(datafile, delimiter=',', usecols=(0,1), unpack=True)
### 코드 끝 ### 

assert cols.shape == (2, 97), "불러온 데이터의 shape가 맞지 않음." 

##############################################################################
### 위 데이터(cols)를 X와 y로 나눈다.                                           ###
### X는 0번째 행을, y는 1번째 행을 나눠가지고 둘 다 transpose를 해준다.              ###
### np.transpose 사용.                                                      ###
##############################################################################

### 코드 시작 ###
X = np.transpose(cols[:1])
y = np.transpose(cols[1:2])
### 코드 끝 ###

assert X.shape == (97, 1) and y.shape == (97, 1), "X 혹은 y의 shape가 맞지 않음."
m = y.size # number of training examples
##############################################################################
### X vector에 값이 1로 이루어진 1개 열을 추가해서 matrix로 만들어준다.              ###
### 강의에서 bias(theta0) 값 X에 넣어주는 연산임.                                ###
### np.insert 사용.                                                        ###
##############################################################################

### 코드 시작 ###
X = np.insert(X, 0, 1, axis=1)
### 코드 끝 ###

assert X.shape == (97, 2), "X의 shape가 맞지 않음"
```

#### 2.2 Gradient Descent

```python
# 반복 횟수
iterations = 1500
# learning_rate
alpha = 0.01
```

```python
# h와 J를 수식으로 구현해보자. 
def h(theta,X): # 강의에서 h식
    """
    theta : [[theta1], [theta0]]로 이루어진 값 (array) 
    X : 입력 데이터에 1로 이루어진 열이 추가된 값 (array)
    """
    ##############################################################################
    ### 강의에서 사용된 h식을 구현하자.(간단함.)                                       ###
    ### X와 theta가 입력되어서 어떻게 계산될지 생각해보면 됨.                            ###
    ### 잘 모르겠으면 손으로 그려보자. 행열 연산은 shape만 생각하면 거의 해결 됨.            ###
    ##############################################################################
    
    ### 코드 시작 ###
    return np.dot(X, theta)
    ### 코드 끝 ###

def computeCost(theta,X,y): # 강의에서 J식(Cost function)
    """
    theta : [[theta1], [theta0]]로 이루어진 값 (array)
    X : 입력 데이터에 1로 이루어진 열이 추가된 값 (array)
    y : label 데이터 (array) 
    """
    ##############################################################################
    ### 강의에서 사용된 Cost function을 구현하자.                                    ###
    ### 식을 그대로 numpy로 구현하면 된다.                                           ###
    ### np.dot, np.sum, np.sqrt 등 다양한 방법으로 구현 가능.                        ###
    ##############################################################################
    
    ### 코드 시작 ###
    return 1/(2*m)*np.sum(np.square(h(theta,X)-y))
    ### 코드 끝 ###

    
##############################################################################
### 초기 theta 값을 지정해주자 (initialization)                                 ###
### np.zeros를 사용하자.                                                      ###
### theta의 shape을 잘 생각해보자. X와 theta가 연산한 뒤에 y가 만들어져야 한다.       ###
##############################################################################

### 코드 시작 ###
initial_theta = np.zeros((2,1))
### 코드 끝 ###
assert initial_theta.shape == (2,1), "theta의 shape이 맞지 않음."

cost = computeCost(initial_theta,X,y)
correct_cost = 32.07273
print("계산한 Cost와 정답과의 차이 : ",rel_error(cost, correct_cost)) 
```

```python
# Gradient descent를 수식으로 구현해보자. 
def descendGradient(X, theta = np.zeros((2,1))):
    """
    theta : [[theta1], [theta0]]로 이루어진 값. (array)
    X : 입력 데이터에 1로 이루어진 열이 추가된 값 (array)  
    """
    jvec = [] # cost 값들 모아서 나중에 plot 그리려고 넣어줌. 
    theta_history = [] # theta의 변화를 plot으로 보려고 넣어줌. 
    
    ##########################################################################################################                             
    ### 약간 어렵겠지만 아래 순서대로 구현해보자.                                                                   ###
    ### 1) theta를 tmp_theta에 넣는다. (계속 바뀌니까 임시 변수 지정)                                               ###
    ### 2) cost를 구하고 jvec에 cost값 추가한다.                                                                 ###
    ### 3) theta_history에 현재 theta값 넣어준다.                                                               ###
    ### 4) tmp_theta --> theta 각각의 값에 대해서 gradient를 계산하고 업데이트 하고 그 값을 theta --> tmp_theta 에 저장 ###
    ### 5) tmp_theta 값을 theta에 지정                                                                       ###
    ### 6) 4~5번 theta 내 값(theta0, theta1, ...)전체에 대해 진행.                                             ###
    ### np.sum, reshape, np.sum 등 사용하면 된다.                                                             ### 
    ### 참고로, j=0, j=1일 때 gradient 식이 다른데 theta0 때문에 X에 1을 넣어줬기 때문에                             ###  
    ### 계산 식을 생각해보면 같게 놓아도 된다. 한 번 써보길 추천.                                                     ###
    ##########################################################################################################
    
    ### 코드 시작 ###
    
    # theta는 계속 변할 값이니까 임시 값(tmp_theta)에 들어갔다가 변했을 때 theta로 들어가도록 해줌. 
    # iteration 만큼 반복 예정. iterations : 1500으로 위에 지정되어 있음 
    for k in range(iterations):
        tmp_theta = theta
        jvec.append(computeCost(theta,X,y))  # Cost 구해서 히스토리 기록
        theta_history.append(theta)   # theta 히스토리 기록 
        # tmp_theta 내에 변수 개수(theta0, theta1) 만큼 반복해서 업데이트 예정 
        for i in range(len(tmp_theta)):
            # 현재 theta 값이랑 tmp_theta 값이랑 같으니까 theta로 gradient를 구하고 tmp_theta를 업데이트하자. 
            tmp_theta[i] = theta[i] - (alpha/m) * np.sum((h(theta,X)-y) * np.array(X[:,i]).reshape(m, 1))
        # 업데이트 된 tmp_theta를 theta에 집어넣는다. 
        theta = tmp_theta
    # 결과 반환 
    return theta, theta_history, jvec
    ### 코드 끝 ### 
```

```python
#Actually run gradient descent to get the best-fit theta values
initial_theta = np.zeros((X.shape[1],1))
theta, theta_history, jvec = descendGradient(X,initial_theta)
correct_theta = np.array([[-3.636063],[1.166989]])
print("계산한 theta와 정답과의 차이 : ", rel_error(theta, correct_theta))

# 위에서 구한 cost를 그래프로 나타낸 것. (따라 해보는 것 추천)
def plotConvergence(jvec):
    plt.figure(figsize=(10,6))
    plt.plot(range(len(jvec)),jvec,'bo')
    plt.grid(True)
    plt.title("Convergence of Cost Function")
    plt.xlabel("Iteration number")
    plt.ylabel("Cost function")
    dummy = plt.xlim([-0.05*iterations,1.05*iterations])
    #dummy = plt.ylim([4,8])


plotConvergence(jvec)
dummy = plt.ylim([4,7])
```

### 3. Linear Regression with multiple variables

변수가 1개 더 추가된 데이터로 해보자.

```python
datafile = 'data/ex1data2.txt'
# 데이터 불러오기 
cols = np.loadtxt(datafile,delimiter=',',usecols=(0,1,2),unpack=True) #Read in comma separated data
# X, y 지정
X = np.transpose(np.array(cols[:-1]))
y = np.transpose(np.array(cols[-1:]))
m = y.size # number of training examples
# X에 1 값으로 이루어진 열 추가 
X = np.insert(X,0,1,axis=1)
```

```python
# 데이터 분포 확인
plt.grid(True)
plt.xlim([-100,5000])
dummy = plt.hist(X[:,0],label = 'col1')
dummy = plt.hist(X[:,1],label = 'col2')
dummy = plt.hist(X[:,2],label = 'col3')
plt.title('Clearly we need feature normalization.')
plt.xlabel('Column Value')
plt.ylabel('Counts')
dummy = plt.legend()
```

```python
# 데이터가 너무 넓게 분포되어 있어서 다루기가 어렵다. 
# 전처리를 해줘야 하는데 보통 평균으로 빼고 표준편차로 나눠준다.

# feature별 평균, 표준편차 저장해준다. 
stored_feature_means, stored_feature_stds = [], []
# X.copy()를 해주는 이유는 그냥 Xnorm = X로 하면 Xnorm이 바뀔 때 또 바뀐다. 원래 데이터를 보존해주기 위함. 
Xnorm = X.copy()
for icol in range(Xnorm.shape[1]):
    stored_feature_means.append(np.mean(Xnorm[:,icol]))
    stored_feature_stds.append(np.std(Xnorm[:,icol]))
    #Skip the first column
    if not icol: continue
    #Faster to not recompute the mean and std again, just used stored values
    Xnorm[:,icol] = (Xnorm[:,icol] - stored_feature_means[-1])/stored_feature_stds[-1]
```

```python
# 전처리한 데이터 분포를 보자. 0을 중심으로 잘 모여있다. 
plt.grid(True)
plt.xlim([-5,5])
dummy = plt.hist(Xnorm[:,0],label = 'col1')
dummy = plt.hist(Xnorm[:,1],label = 'col2')
dummy = plt.hist(Xnorm[:,2],label = 'col3')
plt.title('Feature Normalization Accomplished')
plt.xlabel('Column Value')
plt.ylabel('Counts')
dummy = plt.legend()
```

```python
# gradient descent를 한다. 위에서 작성한 함수를 그대로 쓰면 된다.
initial_theta = np.zeros((Xnorm.shape[1],1))
theta, theta_history, jvec = descendGradient(Xnorm,initial_theta)

# cost의 변화를 그려보자. 
plotConvergence(jvec)
```

```python
# 이제 예측 시간이다. 특정 값에 대해서 가격이 얼마일지 예측해본다. (답은 당연히 없음.)
print("Check of result: What is price of house with 1650 square feet and 3 bedrooms?")
ytest = np.array([1650.,3.])
# 아까 전처리해준거 다시 원복해준다. 
ytestscaled = [(ytest[x]-stored_feature_means[x+1])/stored_feature_stds[x+1] for x in range(len(ytest))]
ytestscaled.insert(0,1)
value_gd = h(theta,ytestscaled) 
theta_gd = theta

# 최종 결과 값이랑 theta 구한거. 
print("$%0.2f" % float(value_gd))
print("theta : \n", theta_gd)
```

```python
from numpy.linalg import inv
# Normal equation을 구현해보자. 우리가 gradient descent로 구한 값과 비슷한 지 확인하기 위함. 

def normEqtn(X,y):
    ##############################################################################
    ### normal equation을 구현하자.                                              ###
    ### 강의에 나오는 식을 참고하면 되고 inv를 사용해야 한다. (inversion matrix)         ###  
    ### 이 식을 돌리면 나오는 값은 theta이다.                                        ### 
    ##############################################################################
    
    ### 코드 시작 ###
    return np.dot(np.dot(np.linalg.inv(np.dot(np.transpose(X),X)), np.transpose(X)), y)
    ### 코드 끝 ###
```

```python
# normal equation으로 구한 theta랑 value 값 확인해보자. 
theta_ne = normEqtn(X,y)
value_ne = h(theta_ne, [1, 1650., 3])

# 최종 결과 값이랑 theta 구한거. 
print("Normal equation prediction for price of house with 1650 square feet and 3 bedrooms")
print("$%0.2f" % float(value_ne))
print("theta : \n", theta_ne)
```

```python
print("normal equation과 gradient descent으로 구한 식의 가격 차이 : ", rel_error(value_gd, value_ne))
```
